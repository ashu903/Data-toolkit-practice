```python
# 3.Explain the following terms.
# The difference in np.array, np.asarray and np.asanyarray.
# The difference between Deep copy and shallow copy.

# ans>> The difference in np.array, np.asarray and np.asanyarray.
# In NumPy, np.array, np.asarray, and np.asanyarray are functions used to create arrays  but they differ in how they handle inputs that are already arrays.
# np.array
# import numpy as np
input_data = [1, 2, 3]
array1 = np.array(input_data)
print("array1:", array1)
existing_array = np.array([4, 5, 6])
array2 = np.array(existing_array)
print("array2:", array2)
print("array2 is existing_array:", array2 is existing_array)

# np.asarray
import numpy as np
input_data = [1, 2, 3]
array1 = np.asarray(input_data)
print("array1:", array1)
existing_array = np.array([4, 5, 6])
array2 = np.asarray(existing_array)
print("array2:", array2)
print("array2 is existing_array:", array2 is existing_array)

# np.asanyarray
import numpy as np
import numpy.ma as ma
input_data = [1, 2, 3]
array1 = np.asanyarray(input_data)

print("array1:", array1)
existing_array = np.array([4, 5, 6])
array2 = np.asanyarray(existing_array)
print("array2:", array2)

print("array2 is existing_array:", array2 is existing_array)
masked_array = ma.masked_array([7, 8, 9], mask=[0, 1, 0])
array3 = np.asanyarray(masked_array)

print("array3:", array3)
# Output: array3: [7 -- 9]
print("array3 is masked_array:", array3 is masked_array)
# Output: array3 is masked_array: True (no new array created)
```

```python
# b.  Overlay a line plot representing the normal distribution's probability density function (PDF).
# ans>>
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Plot the histogram
plt.figure(figsize=(10, 6))  # Optional: Set the figure size for better visibility
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Step 3: Calculate the mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Step 4: Overlay a normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

# Additional: Label the axes and the title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title(f'Histogram of 1000 Samples from a Normal Distribution\nMean: {mean:.2f}, Standard Deviation: {std_dev:.2f}')

# Show the plot
plt.show()

```

```python
# 17.Q Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
# origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
# ans>>
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Create two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.randn(1000)
y = np.random.randn(1000)

# Step 2: Determine quadrants
quadrants = np.where((x >= 0) & (y >= 0), 'Q1',
                     np.where((x < 0) & (y >= 0), 'Q2',
                              np.where((x < 0) & (y < 0), 'Q3', 'Q4')))

# Step 3: Create a DataFrame for easy plotting with Seaborn
data = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Step 4: Plot the scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='tab10', s=50, edgecolor='w', alpha=0.7)

# Step 5: Customize the plot
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.grid(True)

# Show the plot
plt.show()
```

```python
# 18.  With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.
# ans>>
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Step 1: Output to notebook
output_notebook()

# Step 2: Create the data
x = np.linspace(0, 4 * np.pi, 1000)
y = np.sin(x)

# Step 3: Create a figure
p = figure(title='Sine Wave Function', x_axis_label='X Axis', y_axis_label='Y Axis')

# Step 4: Plot the line chart
p.line(x, y, legend_label='Sine Wave', line_width=2)

# Step 5: Customize the plot
p.grid.grid_line_alpha = 0.3

# Show the plot
show(p)
```

```python
# 19.Q Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values,
# label the axes, and set the title as 'Random Categorical Bar Chart
# ans>>
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Step 1: Output to notebook
output_notebook()

# Step 2: Create the data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 101, size=len(categories))

# Step 3: Create a DataFrame
data = pd.DataFrame({'Category': categories, 'Value': values})

# Step 4: Create a ColumnDataSource
source = ColumnDataSource(data)

# Step 5: Create a color mapper
colors = ['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1', '#aea4e3']
cmap = factor_cmap('Category', palette=colors, factors=categories)

# Step 6: Create a figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value')

# Step 7: Plot the bar chart
p.vbar(x='Category', top='Value', width=0.9, source=source, legend_field='Category',
       line_color='white', fill_color=cmap)

# Step 8: Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Step 9: Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

```python
# 20.. Q Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.
# ans>>
import plotly.graph_objs as go
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)
y = np.random.randn(100)

# Create a trace
trace = go.Scatter(x=x, y=y, mode='lines', name='Random Data')

# Create layout
layout = go.Layout(
    title='Simple Line Plot',
    xaxis=dict(title='X Axis'),
    yaxis=dict(title='Y Axis')
)

# Create figure
fig = go.Figure(data=[trace], layout=layout)

# Show plot
fig.show()
```

```python
# 21.Q Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.
# ans>>
import plotly.graph_objs as go
import numpy as np

# Generate random data
labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 101, size=len(labels))

# Create a trace
trace = go.Pie(labels=labels, values=values, hoverinfo='label+percent', textinfo='value+percent', hole=0.3)

# Create layout
layout = go.Layout(title='Interactive Pie Chart')

# Create figure
fig = go.Figure(data=[trace], layout=layout)

# Show plot
fig.show()
```
